```python
import matplotlib.pyplot as plt
 
# Sample data: average delivery time (in hours) by region
delivery_times = {
    "North-East": 22,
    "Mid-West": 35,
    "South-Central": 29,
    "South-West": 52,
    "Pacific-North": 46,
    "Mountain-West": 39,
    "Atlantic-South": 20
}
 
# Color coding based on delivery performance
colors = []
for time in delivery_times.values():
    if time < 24:
        colors.append("green")  # Efficient
    elif 24 <= time <= 48:
        colors.append("orange")  # Moderate
    else:
        colors.append("red")  # Needs improvement
 
# Extracting data for plotting
regions = list(delivery_times.keys())
times = list(delivery_times.values())
 
# Create horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(regions, times, color=colors)
 
# Annotate each bar with delivery time value
for bar, time in zip(bars, times):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             f"{time} hrs", va='center', fontsize=9)
 
# Customize chart
plt.title("Average Package Delivery Time by Region")
plt.xlabel("Average Delivery Time (hours)")
plt.ylabel("Region")
 
# Add custom legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='green', label='< 24 hrs (Efficient)'),
    Patch(facecolor='orange', label='2448 hrs (Moderate)'),
    Patch(facecolor='red', label='> 48 hrs (Needs Improvement)')
]
plt.legend(handles=legend_elements, title="Performance")
 
plt.tight_layout()
plt.show()
```
